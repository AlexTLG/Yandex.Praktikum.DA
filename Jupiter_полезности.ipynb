**Основные горячие клавиши**

* Esc: Переключение между режимом выполнения и редактирования
* Esc + 00 (ноль-ноль): Перезапуск.
* A: Добавление пустой ячейки сверху
* B: Добавление пустой ячейки снизу
* DD: Удаления ячейки
* C: Копирование ячеек
* X: Вырезание ячеек
* V: Вставка ячеек
* Esc + Z: Вернуть удаленную ячейку. Помнит больше одной ячейки))))
